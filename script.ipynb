5. Establish the days of the week when individuals are most active

```python
# First, create a new column with ActivityDate represented as day of the week
activity_df["day_of_week"] = activity_df["ActivityDate"].dt.day_name()
activity_df["day_number"] = activity_df["ActivityDate"].dt.weekday
print(activity_df)
```

```python
## A list with days of the week
week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
```

```python
## Groupby operation
active_minutes_per_day = activity_df.groupby(["day_number"])["VeryActiveMinutes"]
```

```python
## Total Active Minutes
sum_of_active_minutes_per_day = (
    active_minutes_per_day.sum().rename_axis("Day").reset_index(name="Minutes")
)
```

```python
(sum_of_active_minutes_per_day)
```

```python
plt.figure(figsize=(8, 4.21), dpi=100)
plt.title("Total Active Minutes Per Day", pad=20, loc="left")
sns.barplot(data=sum_of_active_minutes_per_day, x=week, y="Minutes")
plt.show()
```

The graph above shows that Tuesday & Wednesday have the most active minutes.

```python
## Average active minutes per day
mean_of_active_minutes_per_day = (
    active_minutes_per_day.mean().rename_axis("Day").reset_index(name="Minutes")
)
print(mean_of_active_minutes_per_day)
```

```python
plt.figure(figsize=(8, 4.21), dpi=100)
plt.title("Average Active Minutes Per Day", pad=20, loc="left")
sns.barplot(data=mean_of_active_minutes_per_day, x=week, y="Minutes")
plt.show()
```

Monday & Tuesday have the highest mean indicating that people are more active on Mondays & Tuesdays on average.


The visualizations above shows that  days of the sum of active minutes are different that the days of the mean of active minutes
To find out the difference, find the count of each day in the dataframe.

```python
daily_counts_active_df = activity_df["day_number"].value_counts()
print(daily_counts_active_df)
```

This shows more people are active on Tuesdays & Wednesdays but people who are active on Mondays & Tuesdays active longer.  In short, This indicates a variation in the count of days and thus the differences in mean.

This means that we cannot say for sure which day are people most active using the sum.

To get a fair conclusion, we should rely on the mean.

Find out how movement (measured by total steps) varies per day

```python
steps_per_day = activity_df.groupby(["day_number"])["TotalSteps"]
average_steps_per_day = (
    steps_per_day.mean().rename_axis("Day").reset_index(name="Steps")
)
print(average_steps_per_day)
```

```python
plt.figure(figsize=(8, 4.21), dpi=100)
plt.title("Average Steps Per Day", pad=20, loc="left")
sns.barplot(data=average_steps_per_day, x=week, y="Steps")
plt.show()
```

This graph shows people walk more on Tuesdays & Saturdays


```python
# calculating total of individual minutes column
very_active_mins = activity_df["VeryActiveMinutes"].sum()
fairly_active_mins = activity_df["FairlyActiveMinutes"].sum()
lightly_active_mins = activity_df["LightlyActiveMinutes"].sum()
sedentary_mins = activity_df["SedentaryMinutes"].sum()
# plotting pie chart
slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
colours = ["green", "yellow", "blue", "pink"]
explode = [0, 0, 0, 0.1]
plt.style.use("default")
plt.pie(slices, labels = labels, 
        colors = colours, wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```
